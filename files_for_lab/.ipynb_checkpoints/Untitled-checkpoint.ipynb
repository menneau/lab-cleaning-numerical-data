# Create a NEW dataframe using only the numerical data called numerical_df. We'll use this for the subsequent questions

# Define a function that differentiates between continuous and discrete variables. Hint: The number of unique values might be useful. Store continuous data into a continuous_df variable and do the same for discrete_df

```python
def split_continuous_discrete(df):
    
    discrete_columns = [col for col in df.columns if df[col].nunique() < 10]
    continuous_columns = [col for col in df.columns if df[col].nunique() > 10]

    return df[continuous_columns], df[discrete_columns]
```

```python
continuous_df, discrete_df = split_continuous_discrete(numerical_df)
```

# Create a function to plot every discrete variable. Do the same with continuous variables.

```python
def plot_discrete_variables(df):

    for col in df.columns:
        sns.countplot(data=df, x=col)
        plt.title(f'Distribution of {col}')
        plt.show()

plot_discrete_variables(discrete_df)
```

```python
def plot_continuous_variables(df):

    for col in df.columns:
        sns.histplot(data=df, x=col, kde=True, bins=30)
        plt.title(f'Distribution of {col}')
        plt.show()

plot_continuous_variables(continuous_df)
```

```python
def plot_outliers(df):

    for col in df.columns:
        sns.boxplot(data=df, x=col)
        plt.title(f'Boxplot for {col}')
        plt.show()

plot_outliers(continuous_df)
```
